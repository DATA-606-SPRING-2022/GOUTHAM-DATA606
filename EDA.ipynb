```python
df_final = pd.DataFrame()
list1 = [0,1,2,38,39]
for i in range(1,12):
  list1.append(((i)*(3))+2)
list_rem = list(range(0,40))
for i in list1:
  if i in list_rem:
    list_rem.remove(i)
for i in range(0,10):
  for j in range(0,11):
    df_i = pd.read_excel(xlsx_files[i][j])
    year = df_i.iat[1,0]
    df_i.insert(38, "Year", year[-4:])
    Crop_variety = df_i.iat[4,0]
    df_i.insert(38, "Crop", Crop_variety[14:])
    df_i = df_i.iloc[6:,:,]
    df_i = df_i.drop(df_i.columns[list_rem],axis=1)
    head_list = df_i.iloc[1,0:2].values.tolist()
    list2 = df_i.iloc[0,2:14].values.tolist()
    list2 = ["Price in "+x for x in list2]
    head_list = head_list + list2
    head_list = head_list + df_i.columns.values.tolist()[-2:]
    df_i.columns = head_list
    df_i = df_i.drop(labels=[6,7],axis = 0)
    df_i= df_i.reset_index(drop= True)
    df_final = df_final.append(df_i, ignore_index = True)

```
